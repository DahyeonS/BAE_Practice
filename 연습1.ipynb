```python
#@title  작업 1유형
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')

df.head()
```

```python
# 1. 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는? (0~9 : 0 , 10~19 : 10)

print((df["age"] // 10 * 10).value_counts().index[0])
```

```python
# 2. 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 나이대 구간의 인원은 몇명인가?

print((df["age"] // 10 * 10).value_counts().iloc[0])
```

```python
# 3. 나이가 25살 이상 29살 미만인 응답 고객들중 housing컬럼의 값이 yes인 고객의 수는?

print(df[(df["age"] >= 25) & (df["age"] < 29) & (df["housing"] == "yes")].shape[0])
```

```python
# 4. numeric한 값을 가지지 않은 컬럼들중 unique한 값을 가장 많이 가지는 컬럼은?

lst = []
for col in df.select_dtypes(exclude = "int"):
  target = df[col]
  lst.append([col, target.nunique()])

print(pd.DataFrame(lst).sort_values(1, ascending = False).iloc[0, 0])
```

```python
# 5. balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID값을 기준으로 내림차순 정렬했을때 상위 100개 데이터의 balance값의 평균은?

print(df[df["balance"] >= df["balance"].mean()].sort_values("ID", ascending = False).head(100)["balance"].mean())
```

```python
# 6. 가장 많은 광고를 집행했던 날짜는 언제인가? (데이터 그대로 일(숫자),달(영문)으로 표기)

print(df[["day", "month"]].value_counts().index[0])
```

```python
# 7. 데이터의 job이 unknown 상태인 고객들의 age 컬럼 값의 정규성을 검정하고자 한다. 샤피로 검정의 p-value값을 구하여라

from scipy import stats
print(stats.shapiro(df["age"][df["job"] == "unknown"])[1])
```

```python
# 8. age와 balance의 상관계수를 구하여라

print(df[["age", "balance"]].corr().iloc[1, 0])
```

```python
# 9. y 변수와 education 변수는 독립인지 카이제곱검정을 통해 확인하려한다. p-value값을 출력하라

from scipy import stats

print(stats.chi2_contingency(pd.crosstab(df["y"],df["education"]))[1])
```

```python
# 10. 각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?

ratio_df = df.groupby(["job", "marital"]).size().reset_index().pivot_table(index = "job", columns = "marital")[0].fillna(0)
ratio_df["ratio"] = ratio_df["divorced"] / ratio_df["married"]

print(ratio_df.sort_values("ratio", ascending = False)["ratio"][0])
```

```python
#@title 작업 2유형
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

display(train.head())
display(test.head())
display(submission.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

y_train = train.pop("y")
test_id = test.pop("ID")

train = train.drop(["day", "month", "ID"], axis = 1)
test = test.drop(["day", "month"], axis = 1)

label = LabelEncoder()
train["job"] = label.fit_transform(train["job"])
train["marital"] = label.fit_transform(train["marital"])
train["education"] = label.fit_transform(train["education"])
train["default"] = label.fit_transform(train["default"])
train["housing"] = label.fit_transform(train["housing"])
train["loan"] = label.fit_transform(train["loan"])
train["contact"] = label.fit_transform(train["contact"])
train["poutcome"] = label.fit_transform(train["poutcome"])
test["job"] = label.fit_transform(test["job"])
test["marital"] = label.fit_transform(test["marital"])
test["education"] = label.fit_transform(test["education"])
test["default"] = label.fit_transform(test["default"])
test["housing"] = label.fit_transform(test["housing"])
test["loan"] = label.fit_transform(test["loan"])
test["contact"] = label.fit_transform(test["contact"])
test["poutcome"] = label.fit_transform(test["poutcome"])
y_train = label.fit_transform(y_train)

scale = MinMaxScaler()
target = ["balance", "pdays"]
train[target] = scale.fit_transform(train[target])

X_tr, X_val, y_tr, y_val = train_test_split(train, y_train, test_size = 0.25, random_state = 1234)

nb = GaussianNB()
nb.fit(X_tr, y_tr)
pred = nb.predict(X_val)
print("Naive Bayes -", roc_auc_score(y_val, pred))

lr = LogisticRegression()
lr.fit(X_tr, y_tr)
pred = lr.predict(X_val)
print("Logistic Regression -", roc_auc_score(y_val, pred))

dt = DecisionTreeClassifier()
dt.fit(X_tr, y_tr)
pred = dt.predict(X_val)
print("Decision Tree -", roc_auc_score(y_val, pred))

rf = RandomForestClassifier(max_depth = 30, n_estimators = 500, min_samples_leaf = 5)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
print("Random Forest -", roc_auc_score(y_val, pred))

xgb = XGBClassifier()
xgb.fit(X_tr, y_tr)
pred = xgb.predict(X_val)
print("XGBoost -", roc_auc_score(y_val, pred))

model = RandomForestClassifier(max_depth = 30, n_estimators = 500, min_samples_leaf = 5)
model.fit(train, y_train)
prediction = model.predict_proba(test)[:,1]

result = pd.DataFrame({"ID":test_id, "predict":prediction})
result.to_csv("result.csv", index = False)
```

```python
# 2. bmi컬럼의 결측치를 bmi컬럼의 결측치를 제외한 나머지 값들의 중앙값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

print(round(df["bmi"].fillna(df["bmi"].median()).mean(), 3))
```

```python
# 3. bmi컬럼의 각 결측치들을 직전의 행의 bmi값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

print(round(df["bmi"].fillna(method = "ffill").mean(), 3))
```

```python
# (4. bmi컬럼의 각 결측치들을 결측치를 가진 환자 나이대(10단위)의 평균 bmi 값으로 대체한 후 대체된 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라)
# Noise 데이터 int형 변환
df['age'] = df['age'].str.replace('*','').astype('int')

# 결측치 제외 나이대별 평균값 계산 및 dictionary 형태로 변환
mean = df[df.bmi.notnull()].groupby(df.age//10 *10).bmi.mean()
dic = { x:y for x,y in mean.items()}


idx =df.loc[df.bmi.isnull(),['age','bmi']].index
# 결측치 매핑
df.loc[df.bmi.isnull(),'bmi']  =(df[df.bmi.isnull()].age//10*10).map(lambda x : dic[x])

result = df.bmi.mean()
print(result)
```

```python
# 5. avg_glucose_level 컬럼의 값이 200이상인 데이터를 모두 199로 변경하고 stroke값이 1인 데이터의 avg_glucose_level값의 평균을 소수점이하 3자리 까지 구하여라

df["avg_glucose_level"][df["avg_glucose_level"] >= 200] = 199

print(round(df["avg_glucose_level"][df["stroke"] == 1].mean(), 3))
```

```python
#@title 작업 2유형

import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/test.csv')

display(train.head())
display(test.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# 데이터셋 불러오기
X_train = train.drop("id", axis = 1)
y_train = X_train.pop("stroke")

X_test = test
test_id = X_test.pop("id")

# 결측값 처리
X_train["bmi"] = X_train["bmi"].fillna(X_train["bmi"].mean())
X_test["bmi"] = X_test["bmi"].fillna(X_test["bmi"].mean())

# 범주형 변수 변환
X_train["age"] = X_train["age"].str.replace("*", "").astype(int)
X_test["age"] = X_test["age"].astype(int)

label = LabelEncoder()

X_train["gender"] = label.fit_transform(X_train["gender"])
X_train["ever_married"] = label.fit_transform(X_train["ever_married"])
X_train["work_type"] = label.fit_transform(X_train["work_type"])
X_train["Residence_type"] = label.fit_transform(X_train["Residence_type"])
X_train["smoking_status"] = label.fit_transform(X_train["smoking_status"])

X_test["gender"] = label.fit_transform(X_test["gender"])
X_test["ever_married"] = label.fit_transform(X_test["ever_married"])
X_test["work_type"] = label.fit_transform(X_test["work_type"])
X_test["Residence_type"] = label.fit_transform(X_test["Residence_type"])
X_test["smoking_status"] = label.fit_transform(X_test["smoking_status"])

# 데이터 정규화
scale = MinMaxScaler()
X_train["avg_glucose_level"] = scale.fit_transform(X_train[["avg_glucose_level"]])

# 데이터셋 분리
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state = 222)

# 모델링
nb = GaussianNB()
nb.fit(X_tr, y_tr)
pred = nb.predict(X_val)
print("Naive Bayes -", roc_auc_score(y_val, pred))

lr = LogisticRegression()
lr.fit(X_tr, y_tr)
pred = lr.predict(X_val)
print("Logistic Regression -", roc_auc_score(y_val, pred))

dt = DecisionTreeClassifier()
dt.fit(X_tr, y_tr)
pred = dt.predict(X_val)
print("Decision Tree -", roc_auc_score(y_val, pred))

rf = RandomForestClassifier()
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
print("Random Forest -", roc_auc_score(y_val, pred))

xgb = XGBClassifier()
xgb.fit(X_tr, y_tr)
pred = xgb.predict(X_val)
print("XGBoost -", roc_auc_score(y_val, pred))

# 최종 모델 선정
model = GaussianNB()
model.fit(X_train, y_train)
prediction = model.predict(X_test)

# 결과물 출력
result = pd.DataFrame({"id":test_id, "stroke":prediction})
result.to_csv("result.csv", index = False)
```

```python
#@title 작업 1유형

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/spotify/spotify.csv')
df.head()
```

```python
# 1. 데이터는 현재 년도별 100곡이 인기순으로 정렬되어 있다. 각 년도별 1~100위의 랭킹을 나타내는 rank컬럼을 만들고 매년도 1위의 bpm컬럼의 평균값을 구하여라

df = df.dropna()

rank = []
for j in range(10):
  for i in range(1, 101):
    rank.append(i)

df["rank"] = rank

print(df["bpm"][df["rank"] == 1].mean())
```

```python
# 2. 2015년도에 가장많은 top100곡을 올린 artist는 누구인가?

print(df[df["top year"] == 2015].value_counts("artist").index[0])
```

```python
# 3. 년도별 rank값이 1~10위 까지의 곡들 중 두번째로 많은 top genre는 무엇인가?

print(df[df["rank"].isin(range(1,11))].value_counts("top genre").index[1])
```

```python
# 4. 피처링의 경우 title에 표시된다. 피처링을 가장 많이 해준 가수는 누구인가?

print(df["title"].str.split("feat. ").str[1].dropna().str[:-1].str.strip().value_counts().index[0])
```

```python
# 5. top year 년도를 기준으로 발매일(year released)과 top100에 진입한 일자 (top year)가 다른 곡의 숫자를 count 했을때 가장 많은 년도는?

print(df[df["top year"] != df["year released"]].value_counts("top year").index[0])
```

```python
# 6. artist 컬럼의 값에 대소문자 상관없이 q 단어가 들어가는 아티스트는 몇명인가?

print(df["artist"][df["artist"].str.lower().str.contains("q")].nunique())
```

```python
# 7. 년도 상관없이 전체데이터에서 1~50위와 51~100위간의 dur 컬럼의 평균값의 차이는?

print(abs(df["dur"][df["rank"].isin(range(1, 51))].mean() - df["dur"][df["rank"].isin(range(51, 101))].mean()))
```

```python
# 8. title을 띄어쓰기 단어로 구분 했을때 가장 많이 나온 단어는 무엇인가? (대소문자 구분 x)

print(df["title"].str.split("\(feat.").str[0].str.lower().str.split().explode().value_counts().index[0])
```

```python
# 9. 년도별 nrgy값의 평균값을 구할때 최대 평균값과 최소 평균값의 차이를 구하여라

mean = df.groupby("top year")["nrgy"].mean().sort_values(ascending = False)
print(mean.iloc[0] - mean.iloc[-1])
```

```python
# 10. artist중 artist type 타입을 여러개 가지고 있는 artist는 누구인가

print(df[["artist", "artist type"]].value_counts().reset_index()["artist"].value_counts().index[0])
```

```python
#@title 작업 2유형

import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/y_train.csv")
test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix

x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/y_train.csv")
test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_test.csv")

x_train = x_train.drop("ID", axis = 1)
y_train = y_train.drop("ID", axis = 1)
test_id = test.pop("ID")

X_tr, X_val, y_tr, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state = 123)

nb = GaussianNB()
nb.fit(X_tr, y_tr)
pred = nb.predict(X_val)
print("Naive Bayes: acc -", accuracy_score(y_val, pred), "auc -", roc_auc_score(y_val, pred))
print(confusion_matrix(y_val, pred))

lr = LogisticRegression()
lr.fit(X_tr, y_tr)
pred = lr.predict(X_val)
print("Logistic Regression: acc -", accuracy_score(y_val, pred), "auc -", roc_auc_score(y_val, pred))
print(confusion_matrix(y_val, pred))

dt = DecisionTreeClassifier()
dt.fit(X_tr, y_tr)
pred = dt.predict(X_val)
print("Decision Tree: acc -", accuracy_score(y_val, pred), "auc -", roc_auc_score(y_val, pred))
print(confusion_matrix(y_val, pred))

rf = RandomForestClassifier()
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)
print("Random Forest: acc -", accuracy_score(y_val, pred), "auc -", roc_auc_score(y_val, pred))
print(confusion_matrix(y_val, pred))

xgb = XGBClassifier()
xgb.fit(X_tr, y_tr)
pred = xgb.predict(X_val)
print("XGBoost: acc -", accuracy_score(y_val, pred), "auc -", roc_auc_score(y_val, pred))
print(confusion_matrix(y_val, pred))

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
prediction = model.predict(test)

result = pd.DataFrame({"ID":test_id, "pose":prediction})
result.to_csv("result.csv", index = False)
```

```python
# 1부터 50까지의 등비수열 합
def func(a, r, n):
  return a * (1 - r ** n) / (1 - r)

func(1, 2, 50)
```

```python
# 피보나치 수열
def fibonacci(n):
    if n <= 1:
        return n

    fib_1 = 0
    fib_2 = 1

    for i in range(2, n+1):
        fib = fib_1 + fib_2
        fib_1 = fib_2
        fib_2 = fib

    return fib

for i in range(10):
  print(fibonacci(i))
```

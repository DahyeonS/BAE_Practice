```python
#@title 1 ~ 10번
import numpy as np
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
# 1. 인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)

print(list(df["channelTitle"].value_counts().head(10).index))
```

```python
# 2. 논란으로 인기동영상이 된 케이스를 확인하고 싶다. dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라

print(list(df["channelTitle"][df["likes"] < df["dislikes"]].unique()))
```

```python
# 3. 채널명을 바꾼 케이스가 있는지 확인하고 싶다. channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라

change = df[["channelId", "channelTitle"]].drop_duplicates().channelId.value_counts()

print(len(change[change > 1]))
```

```python
# 4. 일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?

df["trending_date2"] = pd.to_datetime(df["trending_date2"])
df["dayname"] = df["trending_date2"].dt.day_name()

print(df["categoryId"][df["dayname"] == "Sunday"].value_counts().index[0])
```

```python
# 5. 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라

df["trending_date2"] = pd.to_datetime(df["trending_date2"])
df["dayname"] = df["trending_date2"].dt.day_name()

df.groupby(["dayname", "categoryId"], as_index = False).size().pivot(index='categoryId', columns='dayname')
```

```python
# 6. 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. viewcount대비 댓글수가 가장 높은 영상을 확인하라 (view_count값이 0인 경우는 제외한다)

df["view_count2"] = df["view_count"][df["view_count"] > 0]
df["ratio"] = (df["comment_count"] / df["view_count2"]).dropna()

print(df.sort_values(by = "ratio", ascending = False).iloc[0].title)
```

```python
# 7. 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)

df["view_count"] = df["view_count"][df["view_count"] != 0]
df["ratio"] = (df["comment_count"] / df["view_count"]).dropna()
df["ratio"] = df["ratio"][df["ratio"] != 0]

print(df.sort_values(by = "ratio", ascending = True).iloc[0].title)
```

```python
# 8. like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)

df["likes"] = df["likes"][df["likes"] != 0]
df["dislikes"] = df["dislikes"][df["dislikes"] != 0]

df["ratio"] = (df["dislikes"] / df["likes"]).dropna()

print(df.sort_values(by = "ratio", ascending = True).iloc[0].title)
```

```python
# 9. 가장많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜기준, 중복포함)

print(df["channelTitle"][df["channelId"] == df["channelId"].value_counts().index[0]].unique())
```

```python
# 10. 20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?

print((df[["title", "channelId"]].value_counts() >= 20).sum())
```

```python
# 12. 수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라

video["ct"] = pd.to_datetime(video["ct"])

df = video.sort_values(["videoname", "ct"], ascending = False).drop_duplicates("videoname")
df = df[["videoname", "ct", "viewcnt"]]

df.sort_values("videoname", ascending = True).style.hide_index()
```

```python
# 13. Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라

channel["ct"] = pd.to_datetime(channel["ct"])

df = channel[channel["ct"] >= "2021-10-03"].drop_duplicates("channelname").reset_index(drop = True)
df = df[["channelname", "subcnt"]]

print(df)
```

```python
# 14. 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

start = channel[["channelname", "subcnt"]][channel["ct"].dt.strftime('%Y-%m-%d %H') == "2021-10-03 03"].reset_index(drop = True)
end = channel[["channelname", "subcnt"]][channel["ct"].dt.strftime('%Y-%m-%d %H') == "2021-11-01 15"].reset_index(drop = True)

start.columns = ["channelname", "start"]
end.columns = ["channelname", "end"]

df = pd.merge(start, end)
df["del"] = df["end"] - df["start"]
print(df[["channelname", "del"]])
```

```python
#@title 21 ~ 30번

import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
# 21. 주어진 전체 기간의 각 나라별 골득점수 상위 5개 국가와 그 득점수를 데이터프레임형태로 출력하라

df.groupby("Country").sum().sort_values("Goals", ascending = False).head(5)
```

```python
# 22. 주어진 전체기간동안 골득점을 한 선수가 가장 많은 나라 상위 5개 국가와 그 선수 숫자를 데이터 프레임 형식으로 출력하라

result = pd.DataFrame(df.groupby("Country").size().sort_values(ascending = False).head(5))
print(result)
```

```python
# 23. Years 컬럼은 년도 -년도 형식으로 구성되어있고, 각 년도는 4자리 숫자이다. 년도 표기가 4자리 숫자로 안된 케이스가 존재한다. 해당 건은 몇건인지 출력하라

df['yearLst'] = df.Years.str.split('-')

def checkFour(x):
    for value in x:
        if len(str(value)) != 4:
            return False
        
    return True
    
df['check'] = df['yearLst'].apply(checkFour)

result = len(df[df.check ==False])
result
```

```python
# 24. **Q3에서 발생한 예외 케이스를 제외한 데이터프레임을 df2라고 정의하고 데이터의 행의 숫자를 출력하라 (아래 문제부터는 df2로 풀이하겠습니다) **

df2 = df[df["check"] == True].reset_index(drop = True)
print(df2.shape[0])
```

```python
#@title 104 ~ 113번

import pandas as pd
pd.set_option('display.max_columns',50)
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
df.head()
```

```python
# 104. Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라

print(round(df["Age"][df["Tm"] == "TOR"].mean(), 4))
```

```python
# 105. Pos 컬럼은 포지션을 의미한다. 전체 선수 중 최소나이대의 선수들을 필터하고 그들 중 가장 많은 포지션은 무엇인지 확인하라

print(df["Pos"][df["Age"] == df["Age"].min()].value_counts().index[0])
```

```python
# 106. 선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라

print(df["Player"].str.split().str[0].value_counts().head(1))
```

```python
# 107. PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라

print(df.groupby("Pos")["PTS"].mean())
```

```python
# 108. G컬럼은 참여한 경기의 숫자이다. 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라

print(df.sort_values(["Tm", "G"], ascending = False).groupby("Tm").head(1)["G"].mean())
```

```python
# 109. Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?

print(df["MP"][(df["Tm"] == "MIA") & (df["Pos"].isin(["C", "PF"]))].mean())
```

```python
# 110. 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?

print(df["3P"][df["G"] >= df["G"].mean() * 1.5].mean())
```

```python
# 111. Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?

print(abs(df["G"][df["Age"] >= df["Age"].mean()].mean() - df["G"][df["Age"] <= df["Age"].mean()].mean()))
```

```python
# 112. 평균나이가 가장 젊은 팀은 어디인가

print(df.groupby("Tm")["Age"].mean().sort_values().index[0])
```

```python
# 113. Pos그룹별 평균 MP값을 구하여라

print(df.groupby("Pos")["MP"].mean())
```

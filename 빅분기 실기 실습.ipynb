조건문

```python
a = 10
# a>0는 참이므로 if문 내의 문장을 실행
if (a > 0) :
  print("a는 양의 정수입니다.")
```

```python
# a<0은 거짓이므로 if문 내의 문장을 실행하지 않고, if문 다음 문장 print(a)를 실행
if (a < 0) :
  print("a는 음의 정수입니다.")
print(a)
```

```python
a = -10
# a가 0보다 크면 if 아래 문장을 실행, 작거나 같으면 else 아래 문장을 실행
# a = -10이므로, else 다음 문장인 print("a는 음의 정수입니다.")를 실행
if (a > 0) :
  print("a는 %d입니다." %a)
  print("a는 양의 정수입니다.")
else : 
  print("a는 %d입니다." %a)
  print("a는 음의 정수입니다")
# 문장이 2개 이상일 경우에도 들여쓰기만 정확히 맞춰주면 됨
```

```python
a = -10
# a>0은 거짓이므로 elif 조건인 a==0을 검사하고, 이 조건도 거짓이므로 else 다음 문장을 실행
if (a > 0) :
  print("a는 양의 정수입니다.")
elif (a == 0) :
  print("a는 0입니다.")
else : 
  print("a는 음의 정수입니다.")
```

```python
# 사용자로부터 숫자를 입력 받아서 변수 a, b, c에 대입
a = int(input("첫번째 수를 입력하세요."))
b = int(input("두번째 수를 입력하세요."))
c = int(input("세번째 수를 입력하세요."))

# a와 b를 비교해서 더 큰 값을 big에 대입
if (a >= b) :
  big = a
else :
  big = b

# c와 big을 비교해서 더 큰 값을 출력
if (c >= big) :
  print("가장 큰 수는 %d입니다." %c)
else :
  print("가장 큰 수는 %d입니다." %big)
```

반복문

*for*

```python
for i in range(5) : # 다섯 번을 반복해서 실행
  print("python programming")
```

```python
sum = 0
# range(1, 6)은 1, 2, 3, 4, 5의 5개 숫자를 생성
for i in range(1, 6) : # 변수 i에 1, 2, 3, 4, 5를 대입하면서 반복 실행
  sum = sum + i
  print("i의 값 : %d, 합계 : %d" %(i, sum)) # 2개 이상의 변수를 문자열 포멧팅하기 위해 괄호() 묶음
```

```python
# range(1, 10, 2)는 1부터 시작하여 10까지 2씩 증가하는 숫자(1, 3, 5, 7, 9)를 생성
for i in range(1, 10, 2) :
  print(i, end = ' ') # end = ' '는 결과값을 하나의 줄에서 공백(' ')으로 구분하여 출력하라는 의미
```

```python
for x in "python" : # range 함수 대신에 문자열, 리스트 등도 가능
  print(x) # "python"의 6개 문자를 반복해서 출력
```

```python
# for문 내에 for문 중첩
# i가 1, 2, 3, 4로 반복되는 동안 각각의 i에 대해서 j도 1, 2, 3, 4로 반복됨
for i in range(1, 5) :
  for j in range(1, 5) :
    print(i * j, end = ' ') # print(i * j)는 16번 실행됨
  print() # i가 바뀌는 시점에 줄바꿈
```

```python
for i in range(2, 10) : # i는 2부터 9까지 반복
  for j in range(1, 10) : # j는 1부터 9까지 반복
    print("%d*%d=%d" %(i, j, i * j), end = ' ') # 2*1=2로 시작하여 같은 형태 반복
  print() # 줄바꿈
```

*while*

```python
i = 1
while i <= 5 : # i는 1부터 시작해서 5까지 5번 반복
  print("python programming")
  i = i + 1 # i 값을 증가, 이 구문이 없으면 무한반복됨
```

```python
sum = 0
i = 1
while i <= 5 : # 변수 i 값을 확인하여 반복 실행, i 값이 5가 되었을 때 멈춤
  sum = sum + i
  print("i의 값 : %d, 합계 : %d" %(i, sum)) # 변수 i, sum을 문자열 포멧팅 하기 위해 괄호()로 묶음
  i = i + 1 # 변수 i를 1씩 증가시킴
```

```python
i = 0
s = "python programming"

while i <= (len(s) - 1) : # i가 (문자열 길이 - 1)보다 작을 경우 반복
  print(s[i], end = ' ') # s[i]는 문자열에서 i번째 위치와 문자를 가리킴(문자열 인덱스는 0부터 시작)
  i = i + 1 # 문자열을 구성하는 첫번째 문자부터 마지막 문자까지 한 문자씩 출력
```

```python
i = 0
while i < 100 : # i가 100보다 작은 경우 반복
  print(i, end = ' ') # i의 값 출력
  i = i + 1 # i의 값을 1 증가
  if i == 10 : # i의 값이 10이면 break로 while문을 중지함
    break
```

```python
a = [70, 60, 55, 95, 80, 70, 85, 47, 83, 78]
sum = 0
avg = 0

for x in a :
  sum = sum + x

avg = sum / 10
print("평균은 %f입니다." %avg)
```

```python
a = [70, 60, 55, 95, 80, 70, 85, 47, 83, 78]
sum = 0
avg = 0

i = 0
while i < 10 :
  sum = sum + a[i]
  i = i + 1

avg = sum / 10
print("평균은 %f입니다." %avg)
```

```python
import numpy as np # numpy 패키지를 np라는 이름으로 불러옴

ar1 = np.array([1, 2, 3, 4, 5]) # 리스트 [1, 2, 3, 4, 5]를 ndarray 객체 ar1으로 생성
ar2 = np.array([[1, 2, 3], [4, 5, 6]]) # 2차원 배열 생성
print(ar1) # ar1의 내용 출력
print(ar2) # ar2의 내용 출력
```

```python
import numpy as np # numpy 패키지를 np라는 이름으로 불러옴

ar3 = np.random.randn(3) # 3 크기의 1차원 배열을 생성(난수 발생)
ar4 = np.random.randn(2, 3) # 2*3 크기의 2차원 배열을 생성(난수 발생)
print(ar3)
print(ar4)
```

```python
import numpy as np

ar7 = np.arange(20, 200, 10) # 20부터 200-1까지 10씩 증가
ar8 = ar7.reshape(3, 6) # 3*6 배열로 재배치
print(ar7) # 배열 ar7 출력
print(ar8) # 배열 ar8 출력
```

```python
import numpy as np

ar1 = np.arange(1, 21, 1) # 1부터 20(21-1)까지 1씩 증가
print(ar1) # 배열 ar1 출력
print(ar1[1]) # 배열의 두번째 원소 출력(인덱스는 0부터 시작)

ar2 = ar1.reshape(2, 10) # 2*10 배열로 재구성
print(ar2) # 배열 ar2 출력
ar2[1][1] = 100 # ar2[1][1]의 원소를 100으로 변경
print(ar2[1][0], ar2[1][1], ar2[1][2]) # ar2[1][0], ar2[1][1], ar2[1][2]의 원소 출력
```

```python
ar1 = np.arange(1, 11, 1) # 1부터 10(11-1)까지 1씩 증가
ar2 = ar1 + 3 # ar1 각각의 원소에 3을 더함
ar3 = ar1 * 2 # ar1 각각의 원소에 2를 곱함
print(ar1) # 배열 ar1 출력
print(ar2, "\n", ar3) # 배열 ar2, \n(줄바꿈 기호), ar3 출력
```

```python
ar1 = np.array([[5, 7, 9], [-7, -6, 19], [6, 9, 11]])
print(ar1)
print(ar1.sum()) # 배열 원소의 합계
print(ar1.mean()) # 배열 원소의 평균
print(ar1.max()) # 배열 원소 중 가장 큰 값
print(ar1.min()) # 배열 원소 중 가장 작은 값
print(ar1.max(axis = 0)) # 배열의 각 열의 원소 중 가장 큰 값을 구함
print(ar1.max(axis = 1)) # 배열의 각 행의 원소 중 가장 큰 값을 구함
```

```python
ar1 = np.array([[5, 7, 9], [-7, -6, 19], [6, 9, 11]])
print(ar1 > 0) # ar1의 각 원소가 0보다 큰기 판단한 결과를 출력
m_count = (ar1 < 0).sum() # 원소 중 음수가 몇 개인지 계산(True는 1, False는 0)
print(m_count)

# ar1의 원소가 0보다 작으면 0으로 변경, 그렇지 않으면 ar1의 값으로 대입
ar2 = np.where(ar1 < 0, 0, ar1)
print(ar2)
```

```python
ar1 = np.array([[5, 7, 9], [-3, -6, 19], [6, 4, 11]])
print(ar1) # ar1 출력
ar1.sort(0) # sort(0)는 열 단위로 원소를 정렬
print(ar1)
ar1.sort(1) # sort(!)는 행 단위로 원소를 정렬
print(ar1)
```

*표준정규화*

```python
import numpy as np
import pandas as pd

# 한국인, 일본인 각 성인 1000명 육류소비량 데이터 생성
meat_consumption_korean = 5 * np.random.randn(1000) + 53.9
meat_consumption_japan = 4 * np.random.randn(1000) + 32.7

# 데이터프레임 생성
meat_consumption = pd.DataFrame({"한국인":meat_consumption_korean,
                                 "일본인":meat_consumption_japan})

# 상위 6개 데이터 확인
meat_consumption.head(6)
```

```python
import matplotlib.pyplot as plt

# 한국인 육류소비량 히스토그램
plt.hist(meat_consumption_korean)
plt.xlabel('Korea')
plt.show()

# 일본인 육류소비량 히스토그램
plt.hist(meat_consumption_japan)
plt.xlabel('Japan')
plt.show()
```

*표준정규화 - Z-표준화*

```python
# 표준정규화함수에 의한 변환 zscore() 함수 이용
# 기존 데이터프레임에 새로운 Z-표준화된 칼럼을 추가
import scipy.stats as ss

meat_consumption["한국인_정규화"] = ss.zscore(meat_consumption_korean)
meat_consumption["일본인_정규화"] = ss.zscore(meat_consumption_japan)

meat_consumption.head()
```

```python
# 표준정규화 수식을 직접 입력
# 기존 데이터프레임에 새로운 Z-표준화된 칼럼을 추가

meat_consumption["한국인_정규화2"] = (meat_consumption_korean - np.mean(meat_consumption_korean))/np.std(meat_consumption_korean)
meat_consumption["일본인_정규화2"] = (meat_consumption_japan - np.mean(meat_consumption_japan))/np.std(meat_consumption_japan)

meat_consumption.head()
```

```python
# 표준화된 데이터에 대한 히스토그램
plt.hist(meat_consumption["한국인_정규화"])
plt.xlabel('Korea')
plt.show()

plt.hist(meat_consumption["일본인_정규화"])
plt.xlabel('Japan')
plt.show()
```

```python
# 사이킷런 스케일러 이용 정규화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
meat_consumption["한국인_정규화3"] = scaler.fit_transform(meat_consumption[["한국인"]])
scaler = StandardScaler()
meat_consumption["일본인_정규화3"] = scaler.fit_transform(meat_consumption[["일본인"]])
```

*표준정규화 - Min-Max 정규화*

```python
# Min-Max 변환 MinMaxScaler() 함수 이용
# 기존 데이터프레임에 새로운 표준화된 칼럼을 추가

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
meat_consumption["한국인_mm"] = scaler.fit_transform(meat_consumption[["한국인"]])

scaler = MinMaxScaler()
meat_consumption["일본인_mm"] = scaler.fit_transform(meat_consumption[["일본인"]])

meat_consumption[["한국인", "일본인", "한국인_mm", "일본인_mm"]].head()
```

```python
# Min-Max 변환 수식이용
Min = np.min(meat_consumption_korean)
Max = np.max(meat_consumption_japan)
meat_consumption["한국인_mm2"] = (meat_consumption[["한국인"]] - Min) / (Max - Min)

Min = np.min(meat_consumption_japan)
Max = np.max(meat_consumption_japan)
meat_consumption["일본인_mm2"] = (meat_consumption[["일본인"]]- Min) / (Max - Min)

meat_consumption[["한국인_mm", "일본인_mm", "한국인_mm2", "일본인_mm2"]].head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# apriori. association_rules 모듈 호출
from mlxtend.frequent_patterns import apriori, association_rules
```

*데이터 불러오기*

```python
from google.colab import drive
drive.mount('/content/drive')

# pandas의 read_csv 함수를 통한 csv 데이터 로드
df = pd.read_csv("/content/drive/MyDrive/dataset/빅분기 실습/retail_dataset.csv", sep = ",")
```

*데이터 살펴보기*

```python
df
```

```python
df.info()
```

*데이터 전처리*

```python
# 장바구니 데이터 고유항목 구분 출력
items = set()
for col in df : 
  items.update(df[col].unique())

# 장바구니 목록 값(텍스트)을 수치로 표현 - 각 항복 당 매칭될 경우 1로, 아니면 0으로 표시(one-hot encoding)
itemset = set(items)
encoding = []
for index, row in df.iterrows() :
  rowset = set(row)
  labels = {}
  dismatching = list(itemset - rowset)
  matching = list(itemset.intersection(rowset))
  for i in dismatching : 
    labels[i] = 0
  for j in matching :
    labels[j] = 1
  encoding.append(labels)
encoding[0]
result = pd.DataFrame(encoding)

result
```

*분석 데이터셋 준비*

```python
# 첫째 NaN 열 항목 삭제
result = result.drop(result.columns[0], axis = 1)
result
```

*데이터 분석 수행*

```python
# apriori 함수 적용
freq_items = apriori(result, min_support = 0.2, use_colnames = True)
```

```python
# association_rules로 규칙 도출(신뢰도 임계치 0.6 기반)
rules = association_rules(freq_items, metric = "confidence", min_threshold = 0.6)
rules.head()
```

```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_csv("/content/Bank_Personal_Loan_Modelling.csv")

df = df.drop(df[["ID", "ZIP Code"]], axis = 1)

X = df.drop(["Personal Loan"], axis = 1)
y = df["Personal Loan"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1234)

# Normalize
from sklearn.preprocessing import Normalizer

nor = Normalizer()
X_train = nor.fit_transform(X_train)
X_test = nor.fit_transform(X_test)

from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []

# 1에서 25까지 n_neighbors를 적용
neighbors_settings = range(1, 25)

# KNeighborsClassifier() 함수에 K개수를 입력하여 모델을 만들고 트레이닝 데이터와 테스트 데이터를 적용
for n_neighbors in neighbors_settings :
  # 모델 생성
  clf = KNeighborsClassifier(n_neighbors = n_neighbors)
  clf.fit(X_train, y_train)
  # 훈련 세트 정확도 저장
  training_accuracy.append(clf.score(X_train, y_train))
  # 일반화 정확도 저장
  test_accuracy.append(clf.score(X_test, y_test))

# K의 변화량에 따른 예측정확도 성능 비교
plt.plot(neighbors_settings, training_accuracy, label = "Training Accuracy")
plt.plot(neighbors_settings, test_accuracy, label = "Test Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
test_accuracy = pd.DataFrame(test_accuracy)
k = test_accuracy[0].argmax() + 1
acc = test_accuracy[0].max()
asr = pd.DataFrame({"K값":[k], "정확도":[acc]})
asr.to_csv("111111111.csv", index = False)
```

```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.preprocessing as preprocessing
import statsmodels.api as sm
from sklearn.metrics import f1_score

df = pd.read_csv('/content/Parkinsons.csv')

df = df.drop(["name"], axis = 1)
df1 = df

df = preprocessing.minmax_scale(df)
df = pd.DataFrame(df)
df.columns = df1.columns

df = sm.add_constant(df, has_constant = 'add')

f_columns = list(df.columns.difference(["status"]))
X = df[f_columns]
y = df['status'].astype('category')

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.1, random_state = 2022)
model = sm.Logit(y_train, X_train)
result = model.fit(method = 'bfgs', maxiter = 1000)

def cut_off(y, threshold) :
  Y = y.copy()
  Y[Y > threshold] = 1
  Y[Y <= threshold] = 0
  return (Y.astype(int))

prob = result.predict(X_test)
pred1= cut_off(prob, 0.5)
pred2 = cut_off(prob, 0.8)
score1 = f1_score(y_test, pred1)
score2 = f1_score(y_test, pred2)

ans = pd.DataFrame({'cutoff 0.5': [score1], 'cutoff 0.8':[score2]})
ans.to_csv("111111111.csv", index = False)
```
